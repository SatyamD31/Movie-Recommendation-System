```python
import pandas as pd
import numpy as np
from requests import get
from bs4 import BeautifulSoup
```

```python
import os
from time import sleep
from random import randint
```

```python
pages = np.arange(1, 440282, 250)

title = []
year = []
rating = []
genre = []
director = []
cast = []

headers = {'Accept-Language': 'en-US, en; q=0.5'}
count = 0

for page in pages:
    page = get('https://www.imdb.com/search/title/?title_type=feature&count=250&start=' + str(page) + '&ref_=adv_nxt', headers=headers)
    soup = BeautifulSoup(page.text, 'html.parser')
    siteContent = soup.find(id='main').find_all('div', class_='lister-item mode-advanced')
    count = count+1
    print(count)
    sleep(randint(2, 4))

    for movie in siteContent:
        title.append(movie.h3.a.text)
        year.append(movie.h3.find('span', class_='lister-item-year').text)

        try:
            rating.append(float(movie.strong.text))
        except:
            rating.append(np.nan)
        try:
            genre.append(movie.find('span', class_='genre').text.rstrip().replace('\n','').split(', '))
        except:
            genre.append(['Unavailable'])

        stars = movie.find('p', class_='')
        try:
            casts = stars.text.replace('\n', '').split('|')
            casts = [x.strip('    Director:').strip('    Directors:').strip('     Stars:').strip('Stars:').strip('Star:').strip('Director:') for x in casts]
            casts = [casts[i].replace(j, '') for i, j in enumerate(['', ''])]
            director.append([x.strip('    Director:').strip('    Directors:').strip('     Stars:').strip('Stars:').strip('Star:').strip('Director:')for x in casts[0].split(', ')])
            cast.append([x.strip('    Director:').strip('    Directors:').strip('     Stars:').strip('Stars:').strip('Star:').strip('Director:') for x in casts[1].split(', ')])
        except:
            casts = stars.text.replace('\n', '').strip()
            if(len(casts)>0):
                if(casts[0]=="S"):
                    director.append(['Unavailable'])
                    cast.append([x.strip('    Director:').strip('    Directors:').strip('     Stars:').strip('Stars:').strip('Star:').strip('Director:') for x in casts.split(', ')])
                else:
                    cast.append(['Unavailable'])
                    director.append([x.strip('    Director:').strip('    Directors:').strip('     Stars:').strip('Stars:').strip('Star:').strip('Director:') for x in casts.split(', ')])
            else:
                cast.append(['Unavailable'])
                director.append(['Unavailable'])

        df = pd.DataFrame({'Movie': title, 'Release Year': year, 'IMDB Rating': rating, 'Genre': genre, 'Director': director, 'Stars': cast})
        df.to_csv('Datasets/finalMovieScrap.csv')

        # stop writing to file when file size is greater than 25MB
        if(os.stat('Datasets/finalMovieScrap.csv').st_size > 26214400):
            exit()

print('All data scraped successfully!!')        
```
